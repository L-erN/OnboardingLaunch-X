```python
astVel = 19
adver = "Todos busquen un asteroide en el cielo, ¡la tierra corre peligro!"

if(astVel > 20):
    print(adver)
elif (astVel == 20):
    print(adver)
else:
    print("La tierra no corre peligro aún.")
```
